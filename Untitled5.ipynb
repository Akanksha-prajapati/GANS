```python
# Implement a simple GAN architecture to generate random images (like noise or basic shapes) using TensorFlow/Keras:

import tensorflow as tf
from tensorflow.keras.layers import Dense, Reshape, Flatten, LeakyReLU
from tensorflow.keras.models import Sequential
import numpy as np

# Generator
def build_generator():
    model = Sequential([
        Dense(128, activation="relu", input_shape=(100,)),  # Noise input
        Dense(256, activation="relu"),
        Dense(28*28, activation="tanh"),  # Output image size (flattened)
        Reshape((28, 28, 1))
    ])
    return model

# Discriminator
def build_discriminator():
    model = Sequential([
        Flatten(input_shape=(28, 28, 1)),
        Dense(256),
        LeakyReLU(0.2),
        Dense(128),
        LeakyReLU(0.2),
        Dense(1, activation="sigmoid")
    ])
    return model

generator = build_generator()
discriminator = build_discriminator()

print("Generator Summary:")
generator.summary()
print("\nDiscriminator Summary:")
discriminator.summary()

```

```python
#Train the generator to produce simple digits (using noise as input) and plot the generated images:

import matplotlib.pyplot as plt

def generate_and_plot_images(generator, num_images=5):
    noise = np.random.normal(0, 1, (num_images, 100))
    generated_images = generator.predict(noise)

    fig, axes = plt.subplots(1, num_images, figsize=(10, 2))
    for i in range(num_images):
        axes[i].imshow(generated_images[i].reshape(28, 28), cmap='gray')
        axes[i].axis('off')
    plt.show()

generate_and_plot_images(generator)
```

```python
# Use a trained generator to generate a batch of fake images and display them:

def generate_fake_images(generator, batch_size=10):
    noise = np.random.normal(0, 1, (batch_size, 100))
    generated_images = generator.predict(noise)

    fig, axes = plt.subplots(1, batch_size, figsize=(10, 2))
    for i in range(batch_size):
        axes[i].imshow(generated_images[i].reshape(28, 28), cmap='gray')
        axes[i].axis('off')
    plt.show()

generate_fake_images(generator)
```

```python
# Create a StyleGAN-inspired architecture that outputs high-resolution images:

from tensorflow.keras.layers import Conv2DTranspose, Conv2D, BatchNormalization

def build_stylegan_generator():
    model = Sequential([
        Dense(4*4*512, activation="relu", input_shape=(100,)),
        Reshape((4, 4, 512)),

        Conv2DTranspose(256, kernel_size=4, strides=2, padding="same"),
        BatchNormalization(),
        LeakyReLU(0.2),

        Conv2DTranspose(128, kernel_size=4, strides=2, padding="same"),
        BatchNormalization(),
        LeakyReLU(0.2),

        Conv2DTranspose(64, kernel_size=4, strides=2, padding="same"),
        BatchNormalization(),
        LeakyReLU(0.2),

        Conv2DTranspose(3, kernel_size=4, strides=2, padding="same", activation="tanh")
    ])
    return model

stylegan_generator = build_stylegan_generator()
stylegan_generator.summary()
```

```python
# Write a function to modify the discriminator to include a dropout layer with a rate of 0.4 and print the configurations:

def build_discriminator_with_dropout():
    model = Sequential([
        Flatten(input_shape=(28, 28, 1)),
        Dense(256),
        LeakyReLU(0.2),
        tf.keras.layers.Dropout(0.4),  # Dropout layer
        Dense(128),
        LeakyReLU(0.2),
        tf.keras.layers.Dropout(0.4),  # Dropout layer
        Dense(1, activation="sigmoid")
    ])
    return model

discriminator_dropout = build_discriminator_with_dropout()
discriminator_dropout.summary()
```
